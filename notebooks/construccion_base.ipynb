```python
df_com = pd.read_csv('../../data/EMCALI/lv_data_comercial.csv')
df_com.columns = col_comercial
```

```python
df_com.head()
```

```python
df_com.tipo_producto.value_counts()
```

```python
df_cons = pd.read_csv('../../data/EMCALI/lv_data_consumo.csv')
df_cons.columns = col_consumo
```

```python
df_cons.ano.min(), df_cons.mes.min()
```

```python
df_cons.sort_values(by=['producto', 'ano', 'mes'], inplace=True)
df_cons
```

```python
df_cons['fecha'] = pd.to_datetime(
	df_cons.rename(columns={'ano': 'year', 'mes': 'month'})[['year', 'month']].assign(day=1)
)
```

```python
df = df_cons.merge(df_com[['producto', 'localidad', 'barrio', 'tipo_producto', 'categoria', 'subcategoria']], on='producto', how='left', indicator=True)
df.drop(columns=['lv_fecha_server', 'lv_user'], inplace=True)
```

```python
df._merge.value_counts()
```

```python
df
```

```python
df_sin_comercial = df[df._merge == 'left_only']
df = df[df._merge == 'both']
df.drop(columns=['_merge'], inplace=True)
```

```python
df_sin_comercial.to_csv('../../data/EMCALI/lv_data_consumo_sin_comercial.csv', index=False)
```
